##1) Are movies that are more popular (operationalized as having more ratings) rated higher than movies that are less popular? [Hint: You can do a median-split of popularity to determine high vs. low popularity movies]

```python
#import packages
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#import file as movie data
movie_data = pd.read_csv("/content/movieReplicationSet.csv")
```

```python
#observe data attributes with the .describe() function
movie_data.describe()
```

```python
movies = movie_data.describe().iloc[[0,1,5],:400].T
print(movies)

```

```python
#sort the dataframe by number of ratings
movies = movies.sort_values(by=['count'])
print(movies)
```

```python
# Split the dataframe into two by popularity with the iloc function where .iloc[rows that you want, columns that you want]
low_popularity = movies.iloc[:200,:]
high_popularity = movies.iloc[200:,:]

print("The number of movies in the less popular group is",low_popularity.shape[0])
print("The number of movies in the more popular group is",high_popularity.shape[0])
print("The maximum number of ratings in the less popular group is",low_popularity['count'].max())
print("The median number of ratings of all of the movies is",movies['count'].median())
print("The maximum number of ratings in the less popular group is",high_popularity['count'].min())
```

```python
median_mean_low = low_popularity['mean'].to_numpy()
median_mean_high = high_popularity['mean'].to_numpy()
median_median_low = low_popularity['50%'].to_numpy()
median_median_high = high_popularity['50%'].to_numpy()
```

```python
median_of_means_low = low_popularity['mean'].median()
median_of_means_high = high_popularity['mean'].median()
median_of_medians_low = low_popularity['50%'].median()
median_of_medians_high = high_popularity['50%'].median()


print("The median rating of the means of the low popularity group is",median_of_means_low)
print("The median rating of the means of the high popularity group is",median_of_means_high)
print("The median rating of the medians of the low popularity group is",median_of_medians_low)
print("The median rating of the medians of the low popularity group is",median_of_medians_high)
```

```python
median_of_means_statistic=stats.mannwhitneyu(median_mean_high, median_mean_low, alternative="greater")
median_of_medians_statistic= stats.mannwhitneyu(median_median_high, median_median_low, alternative="greater")
print("The p-value with regard to the median of means is",median_of_means_statistic[1])
print("The p-value with regard to the median of medians is",median_of_medians_statistic[1])
```

Either way, the p-values are well below the alpha of 0.005 and so we can reject the null hypothesis and say that more popular movies are rated higher than less popular movies.

##4) What proportion of movies are rated differently by male and female viewers?

```python
movie_data
```

```python
# code for running a mann whitney u test on each movie to compare male and female ratings
x=[]

for movie_title in movie_data.columns[:400]:

  cleaned_df = movie_data[(movie_data[movie_title].isnull() != True) & (movie_data['Gender identity (1 = female; 2 = male; 3 = self-described)'].isnull() != True) & (movie_data['Gender identity (1 = female; 2 = male; 3 = self-described)'] != 3.0)]
  female_viewers = cleaned_df[(cleaned_df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 1.0)]
  male_viewers = cleaned_df[(cleaned_df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 2.0)]

  #perform mannwhitneyu test
  mwu = stats.mannwhitneyu(female_viewers[movie_title], male_viewers[movie_title], alternative = 'two-sided')
  if mwu[1] < .005:
    x.append(1)
  else:
    x.append(0)

print(x)
```

```python
ratio = x.count(1)/400
print(ratio,"percent of movies are rated differently by men and women!")
```

```python
#above code with running both a mann whitney u and a ks test
y=[]

for movie_title in movie_data.columns[:400]:

  cleaned_df = movie_data[(movie_data[movie_title].isnull() != True) & (movie_data['Gender identity (1 = female; 2 = male; 3 = self-described)'].isnull() != True) & (movie_data['Gender identity (1 = female; 2 = male; 3 = self-described)'] != 3.0)]
  female_viewers = cleaned_df[(cleaned_df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 1.0)]
  male_viewers = cleaned_df[(cleaned_df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 2.0)]

  #perform mannwhitneyu test
  mwu = stats.mannwhitneyu(female_viewers[movie_title], male_viewers[movie_title], alternative = 'two-sided')
  ks = stats.ks_2samp(female_viewers[movie_title], male_viewers[movie_title], alternative = 'two-sided')

  if mwu[1] < .005:
    y.append(1)
  elif ks[1] < .005:
    y.append(1)
  else:
    y.append(0)

print(y)
```

```python
ratio = y.count(1)/400
print(ratio,"percent of movies are rated differently by men and women!")
```

##6) What proportion of movies exhibit an “only child effect”, i.e. are rated different by viewers with siblings vs. those without?



```python
movie_data
```

```python
#running both a mann whitney u and a ks test
y=[]

for movie_title in movie_data.columns[:400]:

  cleaned_df = movie_data[(movie_data[movie_title].isnull() != True) & (movie_data['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] != -1)]
  only_children = cleaned_df[(cleaned_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 1.0)]
  have_siblings = cleaned_df[(cleaned_df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'] == 0.0)]

  #perform mannwhitneyu test
  mwu = stats.mannwhitneyu(only_children[movie_title], have_siblings[movie_title], alternative = 'two-sided')
  ks = stats.ks_2samp(only_children[movie_title], have_siblings[movie_title], alternative = 'two-sided')

  if mwu[1] < .005:
    y.append(1)
  elif ks[1] < .005:
    y.append(1)
  else:
    y.append(0)

print(y)
print(len(y))
```

```python
ratio = y.count(1)/400
print(ratio,'percent of movies exhibit an only child effect!')
```

##8) What proportion of movies exhibit such a “social watching” effect?

```python
#running both a mann whitney u and a ks test
y=[]

for movie_title in movie_data.columns[:400]:

  cleaned_df = movie_data[(movie_data[movie_title].isnull() != True) & (movie_data['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] != -1)]
  lonely = cleaned_df[(cleaned_df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 1.0)]
  social = cleaned_df[(cleaned_df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 0.0)]

  #perform mannwhitneyu test
  mwu = stats.mannwhitneyu(social[movie_title], lonely[movie_title], alternative = 'greater')
  ks = stats.ks_2samp(social[movie_title], lonely[movie_title], alternative = 'greater')

  if mwu[1] < .005:
    y.append(1)
  elif ks[1] < .005:
    y.append(1)
  else:
    y.append(0)

print(y)
print(len(y))
```

```python
ratio = y.count(1)/len(y)
print(ratio,"percent of movies exhibit such a social watching effect, where more social viewers rate a movie higher than viewers who watched alone.")
```

##10) There are ratings on movies from several franchises ([‘Star Wars’, ‘Harry Potter’, ‘The Matrix’, ‘Indiana Jones’, ‘Jurassic Park’, ‘Pirates of the Caribbean’, ‘Toy Story’, ‘Batman’]) in this dataset. How many of these are of inconsistent quality, as experienced by viewers? [Hint: You can use the keywords in quotation marks featured in this question to identify the movies that are part of each franchise]

```python
series_dict = {"Star Wars":[],"Harry Potter":[],"The Matrix":[], 'Indiana Jones':[], 'Jurassic Park':[], 'Pirates of the Caribbean':[], 'Toy Story':[], 'Batman':[]}
for movie in series_dict.keys():
  for col in movie_data.columns:
    if movie in col:
      series_dict[movie].append(col)

print(series_dict)
```

```python
for series_title in series_dict.keys():
  KW = stats.kruskal(*[movie_data[x] for x in series_dict[series_title]], nan_policy = 'omit')
  if KW[1] < .005:
    print('The p-value is',KW[1],'which is below .005. The',series_title,'movies are not of consistent quality.')
  else:
    print('The p-value is',KW[1],'this is greater than/equal to .005. The', series_title,'movies are of consistent quality!')
```
